```python
file_dir = './data'
patient_id = '022318xz'
trial = 4
```

### Format Conversion

`matToCSV()/trcToCSV()/c3dToCSV()`

```python
df = emogis.matToCSV(file_dir, patient_id, trial)
df.head()
```

### Feature Extraction

`motionToJointAngle()`

```python
df = emogis.motionToJointAngle(file_dir, patient_id, trial, save = True)
df.head()
```

### Utility functions

`plot()`

```python
# plot via file_dir, patient_id, trial
emogis.plot(data_type='jnt', 
                file_dir = file_dir, patient_id = patient_id, trial = trial) # default: steps = False, cycle=False
```

```python
# plot via dataframe
emogis.plot(dataframe = df, data_type='jnt') # default: steps = False, cycle=False
```

### Data Processing

`knn_impute()/mice_impute()/interpolate_impute()`

```python
# data imputation via file_dir, patient_id, trial
df_imputed = emogis.knn_impute(file_dir = file_dir, patient_id = patient_id, trial = trial,
                                   data_type = 'jnt') # default: save = False, replace = False
```

```python
# Or, data imputation via dataframe
df_imputed = emogis.knn_impute(dataframe = df, 
                                    data_type = 'jnt', save = False) # , replace = False
```

```python
# if save is true, then either way (dataframe or file_dir) must give file_dir, patient_id, trial
df_imputed = emogis.knn_impute(dataframe = df, 
                                    data_type = 'jnt', 
                                        save = True, replace = True,
                                            file_dir = file_dir, patient_id = patient_id, trial = trial)
```

```python
# Similar for: interpolate_impute, mice_impute, interpolate_impute
```

```python
emogis.plot(dataframe = df_imputed, data_type='jnt')
```

`filter_data()`

```python
df_filtered = emogis.filter_data(dataframe = df_imputed, 
                                    data_type = 'jnt')
```

```python
emogis.plot(dataframe = df_filtered, data_type='jnt')
```
